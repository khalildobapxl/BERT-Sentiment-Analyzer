# Count tokens per text sample
as BERT models have a maximum token limit of 512 tokens, it's important to understand the distribution of text lengths in our dataset. This will help us determine if we need to truncate or pad our sequences during preprocessing.

```python
df_train['word_count'] = df_train['text'].apply(lambda x: len(x.split()))
```
