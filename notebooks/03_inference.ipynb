# Load the trained model

```python
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer

model_path = "../models/imdb-distilbert/checkpoint-3126"
```

```python
classifier = pipeline("sentiment-analysis", model=model_path)
```

```python
print(classifier("I love using transformers library!"))
```

Update model config and tokenizer from pretrained checkpoint to show better output than LABEL_0 and LABEL_1.

```python
model = AutoModelForSequenceClassification.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
model.config.id2label = {0: "NEGATIVE", 1: "POSITIVE"}
model.config.label2id = {"NEGATIVE": 0, "POSITIVE": 1}
```

```python
model.save_pretrained(model_path)
tokenizer.save_pretrained(model_path)
```

```python
classifier = pipeline("sentiment-analysis", model=model_path)
```

```python
print(classifier("I love using transformers library!"))
```
